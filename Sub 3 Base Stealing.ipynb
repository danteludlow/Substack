```python
import pandas as pd
splits = pd.read_csv("C:\\Users\\dante\\Downloads\\running_splits.csv")
feet = ['seconds_since_hit_005', 'seconds_since_hit_010', 'seconds_since_hit_015', 
        'seconds_since_hit_020', 'seconds_since_hit_025', 'seconds_since_hit_030', 'seconds_since_hit_035', 
        'seconds_since_hit_040', 'seconds_since_hit_045', 'seconds_since_hit_050', 'seconds_since_hit_055', 
        'seconds_since_hit_060', 'seconds_since_hit_065', 'seconds_since_hit_070', 'seconds_since_hit_075', 
        'seconds_since_hit_080', 'seconds_since_hit_085', 'seconds_since_hit_090']

#Create correlation matrix and open in excel for easy formatting
correlation_matrix = splits[feet].corr()
correlation_matrix.rename(columns=lambda x: x.replace('seconds_since_hit_', 'feet'), inplace=True)
correlation_matrix.to_csv('correlation_matrix.csv', index=False)
import os
os.system('start excel correlation_matrix.csv')
```

```python
#Splits name reformatting to match FanGraphs 
splits['Name'] = splits[' first_name'] + ' ' + splits['last_name']
splits.drop([' first_name', 'last_name'], axis=1, inplace=True)
splits["Name"] = splits["Name"].str.replace("á", "a")
splits["Name"] = splits["Name"].str.replace("í", "i")
splits["Name"] = splits["Name"].str.replace("ñ", "n")
splits["Name"] = splits["Name"].str.replace("é", "e")
splits["Name"] = splits["Name"].str.replace("ó", "o")
```

```python
teams = pd.read_csv("C:\\Users\\dante\\Downloads\\teams.csv")
teams['SB_A'] = teams['SB'] + teams['CS']
teams['SB_Aplus'] = (teams['SB_A'] / teams['SB_A'].mean()) * 100

teams = teams.rename(columns={'SB': 'team_SB', 'CS': 'team_CS', 'SB_A': 'team_SB_A', 'SB_Aplus': 'team_SB_Aplus'})
teams = teams[['Team', 'team_SB', 'team_CS', 'team_SB_A', 'team_SB_Aplus']]
teams
```

```python
standard = pd.read_csv("C:\\Users\\dante\\Downloads\\standard.csv")
standard = standard[['Name', 'Team', 'SB', 'CS', 'mlbamid']]
standard['SB_A'] = standard['SB'] + standard['CS']
std = standard.merge(teams, on='Team')
std['perc_of_team_att'] = std['SB_A'] / std['team_SB_A']
std
```

```python
advanced = pd.read_csv("C:\\Users\\dante\\Downloads\\advanced.csv")
advanced = advanced[['wSB', 'mlbamid']]
stats = std.merge(advanced, on='mlbamid')
splits = splits.rename(columns={'player_id': 'mlbamid'})
splits = splits[['mlbamid','seconds_since_hit_005','seconds_since_hit_030','seconds_since_hit_090']]
df = stats.merge(splits, on='mlbamid')
df = df.rename(columns={'seconds_since_hit_005': 'feet_005', 'seconds_since_hit_030': 'feet_030', 'seconds_since_hit_090': 'feet_090'})
df
```

```python
# I am realizing I don't have much context for these split times so let's compare them to league average
df['5ft_plus'] = (df['feet_005'] / df['feet_005'].mean()) * 100
df['30ft_plus'] = (df['feet_030'] / df['feet_030'].mean()) * 100
df['90ft_plus'] = (df['feet_090'] / df['feet_090'].mean()) * 100

df_sorted = df.sort_values('wSB', ascending=False)
df_sorted
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df[['team_SB_Aplus', 'perc_of_team_att', '5ft_plus', '30ft_plus', '90ft_plus']]
y = df['wSB']

model = RandomForestRegressor()
model.fit(X, y)

print('Feature Importances:')
feature_importances = model.feature_importances_
for feature, importance in zip(X.columns, feature_importances):
    print(f'{feature}: {importance}')

print('Accuracy Testing:')
from sklearn.metrics import r2_score, mean_squared_error

y_train_pred = model.predict(X)

r2 = r2_score(y, y_train_pred)
print(f"R-squared: {r2}")

mse = mean_squared_error(y, y_train_pred)
print(f"Mean Squared Error: {mse}")
```
