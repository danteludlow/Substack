```python
# Importing and merging savant data

import pandas as pd

change = pd.read_csv('C:\\Users\\dante\\Downloads\\pitch-arsenal-stats.csv')
move = pd.read_csv("C:\\Users\\dante\\Downloads\\pitch_movement.csv")
spin = pd.read_csv("C:\\Users\\dante\\Downloads\\spin-direction-pitches.csv")


columns = ['player_id', 'run_value_per_100', 'run_value', 'pitch_usage', 'hard_hit_percent', 'whiff_percent']
change = change[columns]
columns2 = ['pitcher_id', 'pitch_hand', 'avg_speed', 'pitcher_break_z', 'pitcher_break_x', 'percent_rank_diff_z', 'percent_rank_diff_x']
move = move[columns2]
columns3 = ['player_id', 'spin_rate', 'movement_inches', 'active_spin']
spin = spin[columns3]

merged_df = pd.merge(change, move, left_on='player_id', right_on='pitcher_id')
final_df = pd.merge(merged_df, spin, on='player_id')

mapping = {'R': 1, 'L': 0}
final_df['pitch_hand'] = final_df['pitch_hand'].map(mapping)

final_df
```

```python
# Let's build a random forest!

from sklearn.ensemble import RandomForestRegressor

X = final_df[['pitch_hand', 'avg_speed', 'pitcher_break_x', 'pitcher_break_z', 'spin_rate', 'active_spin']]
y = final_df['run_value_per_100']

model = RandomForestRegressor()
model.fit(X, y)

print('Feature Importances:')
feature_importances = model.feature_importances_
for feature, importance in zip(X.columns, feature_importances):
    print(f'{feature}: {importance}')

print('Accuracy Testing:')
from sklearn.metrics import r2_score, mean_squared_error

y_train_pred = model.predict(X)

r2 = r2_score(y, y_train_pred)
print(f"R-squared: {r2}")

mse = mean_squared_error(y, y_train_pred)
print(f"Mean Squared Error: {mse}")
```

```python
#Adding speed difference

arsenals = pd.read_csv("C:\\Users\\dante\\Downloads\\pitch_arsenals.csv")

velo_columns = ['ff_avg_speed', 'si_avg_speed', 'fc_avg_speed']
arsenals['max_avg_speed'] = arsenals[velo_columns].max(axis=1)

arsenals['ch_velo_dif'] = arsenals['max_avg_speed'] - arsenals['ch_avg_speed']
arsenals_slim = arsenals[['pitcher', 'ch_velo_dif']]

newfinal_df = arsenals_slim.merge(final_df, left_on='pitcher', right_on='player_id')
```
